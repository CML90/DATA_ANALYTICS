```python
#main aspect
# Filtering the DataFrame for countries with missing 'life_exp_at_birth' data
missing_life_exp_countries = df[df['life_exp_at_birth'].isna()]

# Displaying other relevant columns
relevant_columns = ['area', 'birth_rate', 'death_rate', 'infant_mortality_rate', 
           'internet_users', 'life_exp_at_birth', 'maternal_mortality_rate', 
           'net_migration_rate', 'population', 'population_growth_rate']
print(missing_life_exp_countries[relevant_columns])
print("main aspect: low population and area")
```

$\textbf{TASK: Notice population is missing for a few countries, but not the area! Let's use this to our advantage. }$<br>
$\textbf{Fill in the missing population values based on the mean population value for its area.}$

```python

df['population'] = df.groupby('area')['population'].transform(lambda x: x.fillna(x.mean()))
```

$\textbf{TASK:Use the same method for the missing values in birth_rate.}$

```python

df['birth_rate'] = df.groupby('area')['birth_rate'].transform(lambda x: x.fillna(x.mean()))

df = df.dropna(subset=['area'])

median_imputer = SimpleImputer(strategy='median')
mean_imputer = SimpleImputer(strategy='mean')

cols_median_impute = ['infant_mortality_rate', 'maternal_mortality_rate', 'net_migration_rate']
cols_mean_impute = ['birth_rate', 'death_rate', 'internet_users', 'life_exp_at_birth', 'population', 'population_growth_rate']


df[cols_median_impute] = median_imputer.fit_transform(df[cols_median_impute])
df[cols_mean_impute] = mean_imputer.fit_transform(df[cols_mean_impute])

print(df.isna().sum())
```

$\textbf{3. Data Feature Preparation}$

$\textbf{TASK: Create a dummy variable to all categorical data.}$

```python
df_numerical = df.drop('country', axis=1)
```

$\textbf{TASK: For numerical data perform standard scaler.}$

```python
from sklearn.preprocessing import StandardScaler

# Define the scaler
scaler = StandardScaler()

# List of numerical columns (all except 'country')
numerical_cols = ['area', 'birth_rate', 'death_rate', 'infant_mortality_rate',
                  'internet_users', 'life_exp_at_birth', 'maternal_mortality_rate',
                  'net_migration_rate', 'population', 'population_growth_rate']

# Scale the numerical data
df_scaled = scaler.fit_transform(df[numerical_cols])
```

$\textbf{4. Creating and Fitting Kmeans Model}$

$\textbf{TASK:Plot the within sum of squares vs. the number of cluster and find the optimal number of clusters.}$<br>
$\textbf{Fit a k-means model of the optimum cluster.}$

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for i in range(1, 11):  # Testing 1 to 10 clusters
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Fitting K-means with the optimal number of clusters
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans.fit(df_scaled)

# Adding cluster labels to the original DataFrame
df['Cluster_Labels'] = kmeans.labels_
```

$\textbf{5. Model Interpretation}$

$\textbf{TASK: What K value do you think is a good choice?}$<br>
$\textbf{Are there multiple reasonable choices? What features are helping define these cluster choices.}$

I think 3 is good because it has the sharpest bend. However, the curve is still quite smooth and there is no clear elbow bend. Therefore, there are multiple choices: 4 or 5 could work as well. 

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=numerical_cols)
print(centroids)
```

$\textbf{Cluster 0}$ is defined by small area and population (although the population is growing faster than the other clusters), but higher birth,death,infant, and maternity mortality rates. Therefore it has the lowest life expectancy. It has low internet user rates and migration rates.

$\textbf{Cluster 1}$ also has a small area and population, but it also has slow population growth. The four mortality rates are quite low,they have the smallest number of internet users, and have a fairly low migration rate$

$\textbf{Cluster 2}$ have a large area and population but population growth is slowing down. They have low birth,death,infant, and maternity mortality rates, the highest migrate tion, and the most internet users. $
